```python
def SplitTrainAndTest(numberOfTraining, dataFrame):
    msk1=np.random.rand(numberOfTraining)<1
    msk2=np.random.rand(len(dataFrame)-numberOfTraining)<0
    msk=np.concatenate((msk1,msk2), axis=0)
    train=dataFrame[msk]
    test=dataFrame[~msk]
    return train, test
```
